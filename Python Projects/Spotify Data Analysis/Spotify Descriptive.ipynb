```python
#Importing Library and Data
import pandas as pd
pd.set_option('display.max_rows', None)
df = pd.read_csv('spotify_data/streaming_h.csv')


#Data Cleaning- removing time in endTime
df['endTime'] = df['endTime'].str[:10]
```

```python
#Top 10 Artists (With most played songs)
top_10 = df.groupby(by='artistName').count()
top_10 = top_10.sort_values('trackName', ascending=False)
top_10 = top_10.drop(columns=['trackName', 'msPlayed'])
top_10 = top_10.rename(columns={'endTime':'Number of Songs Played'})
top_10.head(10)
```

```python
#Top 10 Songs (By Most Number of Minutes Played)
top_10_songs = df.groupby(['trackName','artistName']).sum('msPlayed')
top_10_songs = top_10_songs.sort_values('msPlayed', ascending=False)
top_10_songs.head(100)
```

```python
#Top 10 Songs (Number of Play Count)
top_10_songs_2 = df.groupby(['trackName','artistName']).count()
top_10_songs_2 = top_10_songs_2.drop(columns=['msPlayed'])
top_10_songs_2 = top_10_songs_2.sort_values('endTime', ascending=False)
top_10_songs_2 = top_10_songs_2.rename(columns={'endTime':'Number of Songs Played'})
top_10_songs_2.head(10)
```

```python
#Top Days which I listened to music
avg_minutes_per_day = df.groupby('endTime').sum('msPlayed')
avg_minutes_per_day['msPlayed']= avg_minutes_per_day['msPlayed'] / 60000
avg_minutes_per_day.sort_values('msPlayed', ascending=False)
```
