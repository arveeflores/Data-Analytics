```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#MSE, smallest MSE is better fit
```

```python
#df_term = pd.read_csv("bank_loan.csv")
df_one = pd.get_dummies(final_dataset["Term"])
```

```python
df_one

# Binary Data is Concatenated into Dataframe
df_two = pd.concat((df_one, final_dataset), axis=1)
 
# Gendercolumn is droped
df_two = df_two.drop(["Term"], axis=1)
 
# We want Short Term =0 and Long Term =1 So we drop Short Term column here
df_two = df_two.drop(["Short Term"], axis=1)

df_two = df_two.drop(["Term"], axis=1)
```

```python
df_two
```

```python
df_two.corr()
```

```python
plt.figure(figsize=(16, 6)) #adjusted the size
heatmap = sns.heatmap(df_two.corr(), vmin=-1, vmax=1, annot =True, cmap="GnBu")
heatmap.set_title("Variable Correlation Heatmap", fontdict ={'fontsize':16},pad = 12)
```

```python
lm = LinearRegression()
X1= df_two [['Current Loan Amount']] #independent variable
Y1 = df_two ['Long Term'] #dependent variable
lm.fit(X1,Y1)
Ypred1 = lm.predict(X1)
mse1 = mean_squared_error(Y1, Ypred1)
mse1
```
